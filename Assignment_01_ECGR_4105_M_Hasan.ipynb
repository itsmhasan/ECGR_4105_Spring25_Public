Problem 1 (50 Points)
Develop a code that runs linear regression with a gradient descent algorithm for each explanatory variable in isolation.
In this case, you assume that in each iteration, only one explanatory variable (either x1, x2, or x3) is explaining the output.
You need to do three different trainings, one per each explanatory variable.
For the learning rate, explore different values between 0.1 and 0.01 (your choice).
Initialize your parameters to zero (theta to zero).

1. Report the linear model you found for each explanatory variable.
2. Plot the final regression model and loss over the iteration per each explanatory variable.
3. Which explanatory variable has the lower loss (cost) for explaining the output (Y)?
4. Based on your training observations, describe the impact of the different learning rates on the final loss and number of training iterations.
    </pre>

```python
# Loading the D3.csv Data
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ECGR_4105/HW/D3.csv')

x1 = data['X1'].values
x2 = data['X2'].values
x3 = data['X3'].values
y = data['Y'].values

m = len(x1)
print(data)
```

```python
# Ploting the Data - Single Variable
import matplotlib.pyplot as plt

plt.scatter(x1, y)
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Scatter Plot of X1 vs Y')
plt.show()

plt.scatter(x2, y)
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('Scatter Plot of X2 vs Y')
plt.show()

plt.scatter(x3, y)
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('Scatter Plot of X3 vs Y')
plt.show()
```

```python
def single_variable_gd(x, y):
  m = len(y)
  lr = 0.01 # Learning Rate
  iters = 100000 #number of iterations

  theta0 = 0.0
  theta1 = 0.0
  coast_history = []

  for i in range(iters):
    h = theta0 + theta1 * x
    j = 1 / (2 * m) * np.sum((h - y)**2) #Calculate coast

    theta0 = theta0 - lr * (1 / m) * np.sum(h - y)
    theta1 = theta1 - lr * (1 / m) * np.sum((h - y)*x)

    coast_history.append(j)
  return theta0, theta1, coast_history

theta0_x1, theta1_x1, coast_history_x1 = single_variable_gd(x1, y)
print(f'theta0: {theta0_x1}, theta1: {theta1_x1}')
print(f"For x1: y ~ {theta0_x1:.4f} + {theta1_x1:.4f} * x1")

plt.plot(range(len(coast_history_x1)), coast_history_x1)
plt.title(" X1 Cost History vs. Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.show()

plt.scatter(x1,y)
plt.plot(x1, (theta0_x1 + theta1_x1 * x1), color='red')
plt.title("x1 vs. y with Regression Line")
plt.xlabel("x1")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
theta0_x2, theta1_x2, coast_history_x2 = single_variable_gd(x2, y)
print(f'theta0: {theta0_x2}, theta1: {theta1_x2}')
print(f"For x2: y ~ {theta0_x2:.4f} + {theta1_x2:.4f} * x2")

plt.plot(range(len(coast_history_x2)), coast_history_x2)
plt.title(" X2 Cost History vs. Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.show()

plt.scatter(x2,y)
plt.plot(x2, (theta0_x2 + theta1_x2 * x2), color='red')
plt.title("x2 vs. y with Regression Line")
plt.xlabel("x2")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
theta0_x3, theta1_x3, coast_history_x3 = single_variable_gd(x3, y)
print(f'theta0: {theta0_x3}, theta1: {theta1_x3}')
print(f"For x3: y ~ {theta0_x3:.4f} + {theta1_x3:.4f} * x3")

# Plot cost history for x3
plt.plot(range(len(coast_history_x3)), coast_history_x3)
plt.title("X3 Cost History vs. Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.show()

# Plot x3 data and the regression line
plt.scatter(x3, y)
plt.plot(x3, (theta0_x3 + theta1_x3 * x3), color='red')
plt.title("x3 vs. y with Regression Line")
plt.xlabel("x3")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

Problem 2 (50 points)
This time, run linear regression with gradient descent algorithm using all three explanatory variables. For the learning rate, explore different values between 0.1 and 0.01 (your choice). Initialize your parameters (theta to zero).

```python
def multivariable_gd(x, y):
  m = len(y)
  lr = 0.001 # Learning Rate
  iters = 100000 #number of iterations

  theta = np.zeros(4)
  coast_history = []

  for i in range(iters):
    h = theta[0] + theta[1] * x[:, 0] + theta[2] * x[:, 1] + theta[3] * x[:, 2]
    j = 1 / (2 * m) * np.sum((h - y)**2) #Calculate coast

    theta[0] = theta[0] - lr * (1 / m) * np.sum((h - y))
    theta[1] = theta[1] - lr * (1 / m) * np.sum((h - y)*x[:, 0])
    theta[2] = theta[2] - lr * (1 / m) * np.sum((h - y)*x[:, 1])
    theta[3] = theta[3] - lr * (1 / m) * np.sum((h - y)*x[:, 2])
    coast_history.append(j)
  return theta, coast_history

x = np.column_stack((x1, x2, x3))
# Train multivariable model
theta_multi, cost_hist_multi = multivariable_gd(x, y)
print("Final multivariable model:")
print(f"y ~ {theta_multi[0]:.4f} + {theta_multi[1]:.4f}*x1 + {theta_multi[2]:.4f}*x2 + {theta_multi[3]:.4f}*x3")


# 1. Plotting Cost History vs. Iterations for the multivariable model
plt.plot(range(len(cost_hist_multi)), cost_hist_multi)
plt.title("Multivariable Cost History vs. Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.grid(True)
plt.show()

# 2. Visualizing Actual vs. Predicted y values
# Compute predictions using the multivariable model:
y_pred_multi = theta_multi[0] + theta_multi[1]*x1 + theta_multi[2]*x2 + theta_multi[3]*x3

plt.scatter(y, y_pred_multi, color='blue')
plt.title("Actual vs. Predicted y (Multivariable)")
plt.xlabel("Actual y")
plt.ylabel("Predicted y")
plt.grid(True)
plt.show()

# Define new input points
new_points = np.array([
    [1, 1, 1],
    [2, 0, 4],
    [3, 2, 1]
])

# Compute predictions for each new point
predictions = theta_multi[0] + theta_multi[1]*new_points[:,0] + theta_multi[2]*new_points[:,1] + theta_multi[3]*new_points[:,2]

# Print the predictions
for point, pred in zip(new_points, predictions):
    print(f"Prediction for (x1, x2, x3) = {point} is y = {pred:.4f}")
```
